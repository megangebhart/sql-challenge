```python
from sqlalchemy import create_engine
from config import pw 
import matplotlib as plt
engine = create_engine(f'postgresql://postgres:{pw}@localhost:5432/sql_homework_v2')
connection = engine.connect()
```

```python
import pandas as pd
salary_df = pd.read_sql('SELECT * FROM salaries', connection)
title_df = pd.read_sql('SELECT * FROM titles', connection)
employees_df = pd.read_sql('SELECT * FROM employees', connection)
```

```python
employee_title_df = title_df.merge(employees_df, left_on = 'title_id', right_on = 'emp_title_id')
```

```python
final_df = employee_title_df.merge(salary_df, on = 'emp_no')
```

```python
salary_df.hist('salary') 
```

```python
#Create a bar chart of average salary by title.
salary_average = final_df.groupby(['title']).mean()['salary']
```

```python
salary_average.plot.bar(x = 'title', y = 'salary')
```
